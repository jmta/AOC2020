```python
from aocd.models import Puzzle
puzzle = Puzzle(year=2020, day=3)
```

```python
#Solve Part 1
if partOne(puzzle.examples[0].input_data) == int(puzzle.examples[0].answer_a):
    print("Think we have an answer for part one")
    puzzle.answer_a = partOne(puzzle.input_data)
else:
    print("Incorrect for part 1")

    
#Solve Part 2
if partTwo(puzzle.examples[0].input_data) == int(puzzle.examples[0].answer_b):
    print("Think we have an answer for part two")
    puzzle.answer_b = partTwo(puzzle.input_data)
else:
    print("Incorrect for part 2")
```

```python
def partOne(input):
    position = [0,0]
    movement = [1,3]
    lines = input.split("\n")
    width = len(lines[0])
    result = 0
    while position[0] < len(lines):
        object = lines[position[0]][position[1]%width]
        if object == '#':
            result += 1
        position[0] = position[0] + movement[0]
        position[1] = position[1] + movement[1]
    return result
```

```python
def partTwo(input):
    movements = [[1,1],[1,3],[1,5],[1,7],[2,1]] #like the last one we just need to change the math and iterate through each movement
    lines = input.split("\n")
    width = len(lines[0])
    result = 1
    for movement in movements:
        position = [0,0]
        hitsThisRun = 0
        while position[0] < len(lines):
            object = lines[position[0]][position[1]%width]
            if object == '#':
                hitsThisRun += 1
            position[0] = position[0] + movement[0]
            position[1] = position[1] + movement[1]
        result = result * hitsThisRun
    print(result)
    return result
```
