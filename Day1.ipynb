```python
from aocd.models import Puzzle
puzzle = Puzzle(year=2020, day=1)
```

```python
#Solve Part 1
if partOne(puzzle.examples[0].input_data) == int(puzzle.examples[0].answer_a):
    print("Think we have an answer for part one")
    puzzle.answer_a = partOne(puzzle.input_data)
else:
    print("Incorrect for part 1")

    
#Solve Part 2
if partTwo(puzzle.examples[0].input_data) == int(puzzle.examples[0].answer_b):
    print("Think we have an answer for part two")
    puzzle.answer_b = partTwo(puzzle.input_data)
else:
    print("Incorrect for part 2")
```

```python
def partOne(input):
    lines = input.split()
    while len(lines) > 0:
        inputA = int(lines.pop())
        listCopy = lines.copy()
        while len(listCopy) > 0:
            inputB = int(listCopy.pop())
            if inputA + inputB == 2020:
                return inputA * inputB
```

```python
def partTwo(input):
    lines = input.split()
    while len(lines) > 0:
        inputA = int(lines.pop())
        listCopy = lines.copy()
        while len(listCopy) > 0:
            inputB = int(listCopy.pop())
            listCopyCopy = listCopy.copy()
            while len(listCopyCopy) > 0:
                inputC = int(listCopyCopy.pop())
                if inputA + inputB + inputC == 2020:
                    return inputA * inputB * inputC
```
