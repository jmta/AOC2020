```python
from aocd.models import Puzzle
puzzle = Puzzle(year=2020, day=2)
```

```python
#Solve Part 1
if partOne(puzzle.examples[0].input_data) == int(puzzle.examples[0].answer_a):
    print("Think we have an answer for part one")
    puzzle.answer_a = partOne(puzzle.input_data)
else:
    print("Incorrect for part 1")

    
#Solve Part 2
if partTwo(puzzle.examples[0].input_data) == int(puzzle.examples[0].answer_b):
    print("Think we have an answer for part two")
    puzzle.answer_b = partTwo(puzzle.input_data)
else:
    print("Incorrect for part 2")
```

```python
def partOne(input):
    lines = input.split("\n")
    result = 0
    while len(lines) > 0:
        inputA = lines.pop()
        lineParts = inputA.split(" ")
        lowerBound = int(lineParts[0].split("-")[0])
        upperBound = int(lineParts[0].split("-")[1])
        timeLetterInPassword = lineParts[2].count(lineParts[1][0])
        if timeLetterInPassword >= lowerBound and timeLetterInPassword <= upperBound:
            result+=1
    return result
        
```

```python
def partTwo(input):
    lines = input.split("\n")
    result = 0
    while len(lines) > 0:
        inputA = lines.pop()
        lineParts = inputA.split(" ")
        firstPosition = int(lineParts[0].split("-")[0])
        secondPosition = int(lineParts[0].split("-")[1])
        letterOfInterest = lineParts[1][0]
        password = lineParts[2]
        if bool(password[firstPosition-1] == letterOfInterest) ^ bool(password[secondPosition-1] == letterOfInterest):
            result+=1
    return result
```
